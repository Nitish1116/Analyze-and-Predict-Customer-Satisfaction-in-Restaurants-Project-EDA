# Data Visualization

```python
data.head()
```

```python
#Correlation between Income and Average Spend
plt.figure(figsize=(10, 6))
sns.histplot(data,x='Income', y='AverageSpend', hue='HighSatisfaction', palette='coolwarm')
plt.title('Income vs Average Spend')
plt.xlabel('Income Rs')
plt.ylabel('Average Spend Rs')
plt.show()
```

```python
#Count of Customers by Preferred Cuisine
plt.figure(figsize=(10, 6))
sns.countplot(data,x='PreferredCuisine',hue='PreferredCuisine', palette='Set2')
plt.title('Count of Customers by Preferred Cuisine')
plt.show()
```

```python
#Distribution of Service Ratings
plt.figure(figsize=(10, 6))
sns.histplot(data['ServiceRating'], bins=5, kde=True, color='green')
plt.title('Distribution of Service Ratings')
plt.show()
```

```python
#Relationship between Wait Time and Service Rating
plt.figure(figsize=(10, 6))
sns.lineplot(data,x='WaitTime', y='ServiceRating', marker='o', hue='HighSatisfaction', palette='muted')
plt.title('Wait Time vs Service Rating')
plt.xlabel('Wait Time')
plt.ylabel('Service Rating')
plt.grid(True)
plt.show()
```

```python
#Gender-based Average Spendings
plt.figure(figsize=(10, 6))
sns.boxplot(data,x='Gender', y='AverageSpend',hue= 'Gender',legend=False,palette='Blues')
plt.title('Gender-wise Average Spendings')
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(10, 6))
correlation = data[['Age', 'Income', 'AverageSpend', 'GroupSize', 'WaitTime', 'ServiceRating', 'FoodRating', 'AmbianceRating']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Customer Count by Visit Frequency
plt.figure(figsize=(10, 6))
sns.countplot(data,x='VisitFrequency',hue='VisitFrequency', palette='Blues')
plt.title('Customer Count by Visit Frequency')
plt.xlabel('Visit Frequency')
plt.ylabel('Count')
plt.show()
```

```python
#Service Rating Distribution by Meal Type
plt.figure(figsize=(10, 6))
sns.violinplot(data,x='MealType', y='ServiceRating',hue='MealType', palette='Set1')
plt.title('Service Rating Distribution by Meal Type')
plt.xlabel('Meal Type')
plt.ylabel('Service Rating')
plt.show()
```

```python
#Average Spend by Dining Occasion
plt.figure(figsize=(10, 6))
sns.barplot(data,x='DiningOccasion', y='AverageSpend',hue='DiningOccasion', palette='Set3')
plt.title('Average Spend by Dining Occasion')
plt.xlabel('Dining Occasion')
plt.ylabel('Average Spend Rs')
plt.show()
```

```python
#Distribution of Food Ratings
plt.figure(figsize=(10, 6))
sns.histplot(data['FoodRating'], bins=5, kde=True, color='purple')
plt.title('Distribution of Food Ratings')
plt.xlabel('Food Rating')
plt.ylabel('Count')
plt.show()
```

```python
#Group Size vs. Wait Time
plt.figure(figsize=(10, 6))
sns.scatterplot(data,x='GroupSize', y='WaitTime', hue='HighSatisfaction', palette='coolwarm', s=100)#s=100: This sets the size of the points to 100, making the points larger for better visibility.
plt.title('Group Size vs. Wait Time')
plt.xlabel('Group Size')
plt.ylabel('Wait Time (Minutes)')
plt.grid(True)
plt.show()
```

```python
#Average Spend by Time of Visit
plt.figure(figsize=(10, 6))
sns.barplot(data,x='TimeOfVisit', y='AverageSpend',hue= 'TimeOfVisit',palette='muted')
plt.title('Average Spend by Time of Visit')
plt.xlabel('Time of Visit')
plt.ylabel('Average Spend Rs')
plt.show()
```

```python
#Average Spend by Preferred Cuisine
plt.figure(figsize=(10, 6))
sns.barplot(data,x='PreferredCuisine', y='AverageSpend',hue='PreferredCuisine', palette='Set2')
plt.title('Average Spend by Preferred Cuisine')
plt.xlabel('Preferred Cuisine')
plt.ylabel('Average Spend Rs')
plt.xticks(rotation=45)
plt.show()
```
